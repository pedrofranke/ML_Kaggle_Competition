```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df_train = pd.read_csv('Datasets/train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('Datasets/test.csv')
df_test.head()
```

```python
plt.subplot(2,1,1)
sns.countplot(data=df_test,x='Gender')
plt.title('Test')
plt.subplot(2,1,2)
sns.countplot(data=df_train,x='Gender')
plt.title('Train')
plt.tight_layout()
plt.show()
```

```python
cat_cols = ['Geography','Gender','IsActiveMember','HasCrCard','NumOfProducts']

for col in cat_cols:
    plt.figure(figsize=(15,8))
    plt.subplot(2,1,1)
    sns.countplot(data=df_train,x=col)
    plt.title(f'{col} in train')
    plt.subplot(2,1,2)
    sns.countplot(data=df_test,x=col)
    plt.title(f'{col} in test')
    plt.tight_layout()
    plt.show()
```

```python
norm_cols = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']

for col in norm_cols:
    plt.figure(figsize=(15,8))
    plt.subplot(2,1,1)
    sns.histplot(data=df_train,x=col)
    plt.title(f'{col} in train')
    plt.subplot(2,1,2)
    sns.histplot(data=df_test,x=col)
    plt.title(f'{col} in test')
    plt.tight_layout()
    plt.show()
```
